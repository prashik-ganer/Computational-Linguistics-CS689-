# **Importing Essentials**

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

access_token = "Add your access token"


tokenizer = AutoTokenizer.from_pretrained(
    "facebook/nllb-200-distilled-600M", token=access_token, src_lang="eng_Latn"
)

model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-600M", token=access_token)

model=model.to("cuda")
```

```python
# Open the file in read mode
with open('/kaggle/input/random-numbers/random_numbers.txt', 'r') as file:
    # Read the lines from the file
    lines = file.readlines()

# Initialize an empty list to store the numbers
random_list = []

# Iterate over each line in the file
for line in lines:
    # Convert the line to an integer and append it to the list
    random_list.append(int(line.strip()))

# Print the list of numbers
# print(random_list)
```

# **ENGLISH TO HINDI**

```python
# Open file B and read the random integers
with open("/kaggle/input/random-numbers/random_numbers.txt", "r") as file_B:
    random_numbers = [int(line.strip()) for line in file_B]

# Open file A and read the lines corresponding to the random integers
with open("/kaggle/input/english/test.en", "r") as file_A:
    # Read all lines from file A into a list
    lines_A = file_A.readlines()

# Print the lines from file A corresponding to the random integers
count=1
with open("output_english_to_hindi.txt", "w", encoding="utf-8") as output_file:
    for index in random_numbers:
        if index < len(lines_A):
            #print(index," ", lines_A[index].strip())
            temp = lines_A[index].strip()
            inputs = tokenizer(temp, return_tensors="pt").to("cuda")
            

            translated_tokens = model.generate(
                **inputs, forced_bos_token_id=tokenizer.lang_code_to_id["hin_Deva"], max_length=30
            )
            result = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]
            print(count, " ",index," ", result)
            count=count+1
            output_file.write(result + "\n")

        
```
