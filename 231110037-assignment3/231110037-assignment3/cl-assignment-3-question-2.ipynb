```python
from nltk.translate.bleu_score import sentence_bleu, corpus_bleu
from rouge import Rouge

def calculate_scores(references, hypotheses):
    bleu_score = corpus_bleu([[ref] for ref in references], hypotheses)
    rouge = Rouge()
    rouge_scores = rouge.get_scores(hypotheses, references, avg=True)
    return bleu_score, rouge_scores
```

# **ROUGE and BLEU scores for NLLB Model**

```python
BLEU_en_hi, ROUGE_en_hi = calculate_scores(original_random_outputs_hindi, nllb_output_english_to_hindi)
BLEU_hi_en, ROUGE_hi_en = calculate_scores(original_random_outputs_english, nllb_output_hindi_to_english)
BLEU_hi_mr, ROUGE_hi_mr = calculate_scores(original_random_outputs_marathi, nllb_output_hindi_to_marathi)
BLEU_mr_hi, ROUGE_mr_hi = calculate_scores(original_random_outputs_hindi, nllb_output_marathi_to_hindi)
print("BLEU, ROUGE scores of ENGLISH TO HINDI translation", BLEU_en_hi, ROUGE_en_hi, "\n") 
print("BLEU, ROUGE scores of HINDI TO ENGLISH translation", BLEU_hi_en, ROUGE_hi_en, "\n")
print("BLEU, ROUGE scores of HINDI TO MARATHI translation", BLEU_hi_mr, ROUGE_hi_mr, "\n")
print("BLEU, ROUGE scores of MARATHI TO HINDI TRANSLATION", BLEU_mr_hi, ROUGE_mr_hi, "\n")

```
