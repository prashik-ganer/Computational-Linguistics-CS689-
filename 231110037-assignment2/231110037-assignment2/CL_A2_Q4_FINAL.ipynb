**Generating the list of keys for occurances of a separate class**

```python
def CountOccurances(ANSWERS, className):
  list_temp=[]
  for sublist in ANSWERS:
    for i in sublist:
      temp = i.split(':')

      if(temp[1]==className):
        list_temp.append(temp[0])

  return list_temp


# def CountOccurances(ANSWERS, className):
#   list_PER=[]
#   for sublist in ANSWERS:
#     for i in sublist:
#       temp = i.split(':')

#       if(temp[1]=='B-'+className or temp[1]=='I-'+className or temp[1]==className):
#         list_PER.append(temp[0])
#   return list_PER
```

**Calculating Precision, Recall and F1 score**

```python
def calculate_metrics(true_labels, predicted_labels):
    true_positives = len(set(true_labels)&set(predicted_labels))
    false_positives = len(set(true_labels)-set(predicted_labels))
    false_negatives = len(set(predicted_labels)-set(true_labels))

    precision = true_positives / (true_positives + false_positives) if (true_positives + false_positives) != 0 else 0
    recall = true_positives / (true_positives + false_negatives) if (true_positives + false_negatives) != 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0

    return precision, recall, f1_score
```
